```python
## task 1
import pandas as pd

file_path = 'GlobalLandTemperaturesByCountry.csv' 
df = pd.read_csv(file_path)

print(df.head(10))

print(df.info())

print(df.describe())
```

```python
## task 2
df.isnull()
```

```python
missing_data = df.isnull().sum()
print(missing_data)
```

```python
df2 = df.fillna(value = 0)
df2
```

```python
df3 = df.dropna(how='any')
df3
```

```python
## task 3
date_col = pd.DatetimeIndex(df3['dt'])
df3.loc[:, 'year'] = date_col.year
df3.loc[:, 'month'] = date_col.month
df3
```

```python
## task 4a
threshold = 2.0
df4 = df[df['AverageTemperatureUncertainty'] <= threshold]
df4
```

```python
original_size = df.shape[0]
print(f"Original dataset size: {original_size}")
filtered_size = df4.shape[0]
print(f"Filtered dataset size: {filtered_size}")
```

```python
## task 4b
df4b = df3.groupby('year').agg({
    'AverageTemperature': 'mean',
    'AverageTemperatureUncertainty': 'mean'
}).reset_index()
df4b
```

```python

import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(20, 10))

years = df4b['year']
avg_temp = df4b['AverageTemperature']
uncertainty = df4b['AverageTemperatureUncertainty']
plt.errorbar(years, avg_temp, yerr=uncertainty, fmt='-o', ecolor='red', capsize=2, label='Average Temperature ± Uncertainty')

threshold = 2.0
high_uncertainty_years = df4b[df4b['AverageTemperatureUncertainty'] > threshold]
plt.scatter(high_uncertainty_years['year'], high_uncertainty_years['AverageTemperature'], 
            color='green', label='High Uncertainty (> 2.0)', zorder=5)
plt.title('Average Temperature Trend with Uncertainty', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Average Temperature (°C)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()

plt.show()
```

```python
## task 4c
df4c = df4b[['year', 'AverageTemperatureUncertainty']].copy()
plt.figure(figsize=(20, 10))

years = df4c['year']
mean_uncertainty = df4c['AverageTemperatureUncertainty']

plt.plot(years, mean_uncertainty, '-o', color='blue', label='Mean Temperature Uncertainty')
plt.title('Trend of Average Temperature Uncertainty Over Time', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Mean Uncertainty (°C)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()

plt.show()
```

```python
## task 4d
threshold = 2
df4d = df4b[df4b['AverageTemperatureUncertainty'] > threshold]
print("Years with High Uncertainty:")
print(df4d)
```
